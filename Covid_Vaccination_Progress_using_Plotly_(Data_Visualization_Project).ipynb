```python
#Showing the null values as per attributes
null_columns=df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```

```python
##Filling null values of daily_vaccinations using mode
df['daily_vaccinations'].fillna(df['daily_vaccinations'].mode()[0], inplace=True)
#Finding the median of the total immunizations
df['total_immunizations'] = df['total_immunizations'].fillna(df['total_immunizations'].median())
#Finding the median of the people_vaccinated
df['people_vaccinated'] = df['people_vaccinated'].fillna(df['people_vaccinated'].median())
#people_fully_vaccinated_per_hundred using median
df['people_fully_vaccinated_per_hundred'] = df['people_fully_vaccinated_per_hundred'].fillna(df['people_fully_vaccinated_per_hundred'].median())
#Finding the median of the people_vaccinated daily_vaccinations_per_hundred
df['total_vaccinations_per_hundred'] = df['total_vaccinations_per_hundred'].fillna(df.groupby('total_immunizations')['total_vaccinations_per_hundred'].transform('median'))
##Finding the median of the people_vaccinated daily_vaccinations_per_million
df['daily_vaccinations_per_million'] = df['daily_vaccinations_per_million'].fillna(df.groupby('daily_vaccinations')['daily_vaccinations_per_million'].transform('median'))
df.head()
```

```python
#people_vaccinated_per_hundred
df['people_fully_vaccinated_per_hundred'] = df['people_fully_vaccinated_per_hundred'].fillna(df['people_fully_vaccinated_per_hundred'].median())
```

```python
#Converting to int
df.total_immunizations = df.total_immunizations.astype('int64') 
df.people_vaccinated = df.people_vaccinated.astype('int64') 
df.	daily_vaccinations = df.daily_vaccinations.astype('int64') 
df.total_vaccinations_per_hundred = df.total_vaccinations_per_hundred.astype('int64') 
df.daily_vaccinations_per_million = df.daily_vaccinations_per_million.astype('int64') 
df.people_fully_vaccinated_per_hundred=df.people_fully_vaccinated_per_hundred.astype('int64')
df.head()
```

```python
def plot_time_variation_countries_group(df, feature, title,Southern_asia):
    data = []
    for country in Southern_asia:
        df2 = df.loc[df.Country==country]
        trace = go.Scatter(
            x = df2['Date'],y = df2[feature],
            name=country,
            mode = "markers+lines",
            marker_line_width = 1,
            marker_size = 8,
            marker_symbol = 'circle',
            text=df2['Country'])
        data.append(trace)
    layout = dict(title = title,
          xaxis = dict(title = 'Date', showticklabels=True,zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='black', mirror=True,
                       tickfont=dict(size=10,color='darkblue'),), 
          yaxis = dict(title = feature, gridcolor='lightgrey', zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='black', mirror=True, type="log"),
                       plot_bgcolor = 'rgba(0, 0, 0, 0)', paper_bgcolor = 'rgba(0, 0, 0, 0)',
         hovermode = 'x',
         height=400
         )
    fig = dict(data=data, layout=layout)
    iplot(fig, filename='all_countries')
```

```python
def plot_time_variation_countries_group(df, feature, title,Africa):
    data = []
    for country in Africa:
        df1 = df.loc[df.Country==country]
        trace = go.Scatter(
            x = df1['Date'],y = df1[feature],
            name=country,
            mode = "markers+lines",
            marker_line_width = 1,
            marker_size = 8,
            marker_symbol = 'circle',
            text=df1['Country'])
        data.append(trace)
    layout = dict(title = title,
          xaxis = dict(title = 'Date', showticklabels=True,zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='black', mirror=True,
                       tickfont=dict(size=10,color='darkblue'),), 
          yaxis = dict(title = feature, gridcolor='lightgrey', zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='black', mirror=True, type="log"),
                       plot_bgcolor = 'rgba(0, 0, 0, 0)', paper_bgcolor = 'rgba(0, 0, 0, 0)',
         hovermode = 'x',
         height=400
         )
    fig = dict(data=data, layout=layout)
    iplot(fig, filename='all_countries')
```

```python
def draw_trace_bar(data, feature, title, xlab, ylab,color='Blue'):
    data = data.sort_values(feature, ascending=False)
    trace = go.Bar(
            x = data['Country'],
            y = data[feature],
            marker=dict(color=color),
            text=data['Country']
        )
    data = [trace]

    layout = dict(title = title,
              xaxis = dict(title = xlab, showticklabels=True, tickangle=45, 
                           zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                           showline=True, linewidth=2, linecolor='black', mirror=True,
                          tickfont=dict(
                            size=10, color='black'),), 
              yaxis = dict(title = ylab, gridcolor='lightgrey', zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                          showline=True, linewidth=2, linecolor='black', mirror=True),
              plot_bgcolor = 'rgba(0, 0, 0, 0)', paper_bgcolor = 'rgba(0, 0, 0, 0)',
              hovermode = 'closest'
             )
    fig = dict(data = data, layout = layout)
    iplot(fig, filename='draw_trace')
```
