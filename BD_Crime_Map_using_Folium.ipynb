```python
crime = {'Name': ['DMP','CMP','KMP','RMP','SMP','GMP','BMP','MR','RPMP'],
        
        'Lon': [23.81,22.36,22.80,24.43,24.89,24.01,22.70,24.74,25.74],
        'Lat': [90.352457,91.82,89.54,88.54,91.93,90.41,90.34,90.41,89.27],
        'Year':[2019,2019,2019,2019,2019,2019,2019,2019,2019],
        'Total Cases':[2153,688,317,401,143,244,166,1004,121],
         'Speedy Trial':[10,5,1,1,3,1,0,1,0],
         'Decocity':[1,0,1,0,0,2,0,1,0],
         'Murder':[19,9,4,1,5,3,0,23,1],
         'Woman & Child Repression':[118,22,8,11,15,22,14,87,12],
         	'Kidnapping':[9,0,0,0,0,1,1,2,1],
         'Police Assault':[6,0,0,0,13,2,0,6,0],
         'Arms Act - RC':[6,38,2,3,0,3,1,5,0],
         'Explosive - RC':[3,0,0,0,0,0,0,2,0],
         'Narcotics - RC':[1370,449,272,283,44,130,91,435,68],
         'Smuggling - RC':[14,1,3,35,2,2,0,7,0],
         'Total - RC':[1393,488,277,321,46,135,92,449,68]


         }
        
dff = pd.DataFrame(crime, columns = ['Name','Lon','Lat','Year','Total Cases','Speedy Trial','Decocity','Murder','Woman & Child Repression','Kidnapping','Police Assault', 'Arms Act - RC','Explosive - RC','Narcotics - RC','Smuggling - RC','Total - RC'])
```

```python
dff
```

```python
locations = dff[['Lon','Lat']]

# Create `deaths_list` by transforming the DataFrame to list of lists 
d_list = locations.values.tolist()

# Check the length of the list
len(d_list)
```

```python
import folium
Fig2=Figure(width=800,height=400)


map = folium.Map(location=[24,90], tiles='Stamen Toner',min_zoom=4,max_zoom=17,zoom_start=14)
Fig2.add_child(map)
for point in range(0, len(d_list)):
    folium.CircleMarker(d_list[point],popup=dff['Name'][point] ,radius=10, color='red', fill=True, fill_color='red', opacity = 0.7).add_to(map)
    folium.TileLayer('Stamen Toner').add_to(map)
folium.TileLayer('Stamen Water Color').add_to(map)
folium.TileLayer('cartodbpositron').add_to(map)
folium.TileLayer('cartodbdark_matter').add_to(map)
folium.LayerControl().add_to(map)
map
```

```python
mapFname = 'output.html'
map.save(mapFname)
```

```python
dff['Total Cases'] = dff['Total Cases'].astype(float)
```

```python
# Make an empty map
Fig3=Figure(width=800,height=400)
mc = folium.Map(location=[24,90], min_zoom=4,max_zoom=17,zoom_start=14,tiles='cartodbpositron')
Fig3.add_child(mc)


# I can add marker one by one on the map
for i in range(0,len(dff)):
   folium.Circle(
      location=[dff.iloc[i]['Lon'],dff.iloc[i]['Lat']],
      popup=dff.iloc[i][['Name','Year','Total Cases','Speedy Trial']],
      radius=dff.iloc[i]['Total Cases']*5,
      color='red',
      fill=True,
      fill_color='red'
   ).add_to(mc)
mc
```

```python
mapFname = 'output.html'
mc.save(mapFname)
```
